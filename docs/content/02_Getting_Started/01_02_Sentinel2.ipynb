# Sentinel-2 L2A

## Product description

The Sentinel-2 L2A data product available in the SALDi Data Cube (SDC) is a 
mirror of [Digital Earth Africa](https://www.digitalearthafrica.org/)'s product 
of the same name. Detailed information can be found [here](https://registry.opendata.aws/deafrica-sentinel-2/).

## Loading and exploring

```python
from sdc.load import load_product

ds = load_product(product="s2_l2a", 
                  vec="site06", 
                  time_range=("2020-04-01", "2020-06-01"),
                  apply_mask=True)
ds
```
